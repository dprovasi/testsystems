```python
def inspect(label, gate_cutoff, md_mg_cutoff, mg_state="md_mg"):
    
    
    for i,tt in enumerate(macro_occup[label]):
        print(f'{i} -> total bound {sum(tt)}, fraction {sum(tt)/len(tt)}')
    
    ## join all cvs
    cvs = [ np.append(cv_real[label][i], 
                      macro_occup[label][i].reshape(len(macro_occup[label][i]),1),1)
           for i in range(len(cv_real[label]))]

    b_list = [pd.DataFrame(trj,columns=['gate','md_na','md_mg','occ_mg']) for trj in cvs]
    b = pd.concat(b_list, keys=range(len(cv_real[label])),names=['trjID'])
    
    
    
    b['md_mg'].hist(by=b['occ_mg'])
    plt.show()
    
    b['gate'].plot.hist(bins=40,alpha=.5)
    plt.axvline(x=gate_cutoff,color='red')
    plt.show()
    
    b[::10].plot.hexbin(x='gate',y='md_mg',gridsize=15,cmap="viridis")
    plt.axvline(x=gate_cutoff,color='red')
    plt.axhline(y=md_mg_cutoff,color='pink')

    
    b['dgate']=(b['gate']<gate_cutoff).astype('int')
    b['dmd_mg']=(b['md_mg']<md_mg_cutoff).astype('int')
    
    if mg_state == 'md_mg':
        b['microstate'] = b['dgate'] + 2*b['dmd_mg']
    else:
        b['microstate'] = b['dgate'] + 2*(b['occ_mg']==1)

    
    dtr = [np.array(grp['microstate']) for name, grp  in b.groupby('trjID')]

    
    return b, dtr
```

```python
lags = [1,5,10,20,50,100,200]

def tscales(dtraj, lags=lags, HMM_states=4, doMSM=False, doHMM=True):
    '''
    plots implied timescales for MSM or HMM
    '''
    if doMSM:
        msm_ts = pyemma.msm.its(dtraj, lags=lags, errors='bayes')
        pyemma.plots.plot_implied_timescales(msm_ts)
        plt.show()
    if doHMM:    
        its_hmm = pyemma.msm.timescales_hmsm(dtraj_ina, HMM_states, lags=lags, errors='bayes')
        pyemma.plots.plot_implied_timescales(its_hmm, ylog=True)
        plt.show()
```

```python
def model1(pis, M):
    '''
    calculates the Js and the Hs for a given system
    '''
    return {
        'JO' : 1./M* pis[2]/pis[0],
        'JC' : 1./M* pis[3]/pis[1],
        'H0' : pis[0]/pis[1],
        'HM' : pis[2]/pis[3] }
```

```python
def checkmd(model_unbound, model_bound, M):
    
    iH0B = 1./model_bound['H0']
    iH0U = 1./model_unbound['H0']
    JCB = model_bound['JC']
    JCU = model_unbound['JC']
    JOB = model_bound['JO']
    JOU = model_unbound['JO']
    
    M50 = (1+ iH0B)/(JOB + iH0B*JCB)
    rho0 = (1+iH0U)/(1+iH0B)
    rhoInf2 = (1+iH0U*JCU/JOU)/(1+iH0B*JCB/JOB)

    
    iHU = iH0U* (JCU*M + 1.)/(JOU*M + 1.)
    iHB = iH0B* (JCB*M + 1.)/(JOB*M + 1.)
    rho = (1+iHU)/(1+iHB)

    plt.plot(np.log10(M),rho,linewidth=3)
    plt.axvline(x=np.log10(M50),color='red', linestyle='dashed')
    plt.axhline(y=rho0,color='purple', linestyle='dashed')
    plt.axhline(y=rhoInf2,color='purple', linestyle='dashed')
    
    zetaU = JCU/JOU
    zetaB = JCB/JOB
    
    
    return {'alpha' :  iH0B/iH0U, 
            'beta': JCB/JCU, 
            'betaO': JOB/JOU, 
            'M50': M50,
            'rho0': rho0,
            'rhoInf2':rhoInf2,
            'zetaU' : zetaU,
            'zetaB' : zetaB
           }
    
def rho(model_unbound, model_bound,M, ):
    
    iH0B = 1./model_bound['H0']
    iH0U = 1./model_unbound['H0']
    JCB = model_bound['JC']
    JCU = model_unbound['JC']
    JOB = model_bound['JO']
    JOU = model_unbound['JO']
    
    iHU = iH0U* (JCU*M + 1.)/(JOU*M + 1.)
    iHB = iH0B* (JCB*M + 1.)/(JOB*M + 1.)
    rho = (1+iHU)/(1+iHB)
    
    
    return rho
```
